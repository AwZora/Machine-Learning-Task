# Cross Validation

Cross-validation is a technique used to evaluate machine learning models more reliably.

Instead of using a single train-test split, the dataset is divided into multiple folds. The model is trained and evaluated several times using different subsets of the data.

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

iris = load_iris()

logreg = LogisticRegression(
    max_iter=1000
)

scores = cross_val_score(
    logreg,
    iris.data,
    iris.target,
    cv=5
)

print("Cross-validation scores:")
print(scores)
```

```python
print("Average cross-validation score: {:.3f}".format(
    scores.mean()
))
```

### Discussion

Cross-validation provides a more robust estimate of model performance because the model is evaluated on multiple train-test splits.

This reduces the risk of obtaining misleading results from a single split.

# K-Fold Cross Validation

K-Fold Cross Validation divides the dataset into K equally sized folds.

Each fold is used once as a test set while the remaining folds are used for training.

```python
from sklearn.model_selection import KFold

kfold = KFold(
    n_splits=5
)

print(kfold)
```

```python
scores = cross_val_score(
    logreg,
    iris.data,
    iris.target,
    cv=kfold
)

print(scores)
```

### Discussion

K-Fold validation ensures that all observations participate in both training and testing, leading to more reliable performance estimates.

# Leave-One-Out Cross Validation

Leave-One-Out Cross Validation (LOOCV) uses a single observation as the test set and the remaining observations as the training set.

This process is repeated for every sample in the dataset.

```python
from sklearn.model_selection import LeaveOneOut

loo = LeaveOneOut()

scores = cross_val_score(
    logreg,
    iris.data,
    iris.target,
    cv=loo
)

print("Number of iterations:",
      len(scores))
```

### Discussion

LOOCV uses almost all available data for training but can be computationally expensive for large datasets.

# Grid Search

Machine learning models often contain parameters that must be specified before training.

Grid Search is a systematic method for testing multiple parameter combinations and selecting the best configuration.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

param_grid = {
    "C": [0.1, 1, 10, 100],
    "gamma": [0.001, 0.01, 0.1, 1]
}
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris.data,
    iris.target,
    random_state=0
)

grid_search = GridSearchCV(
    SVC(),
    param_grid,
    cv=5
)

grid_search.fit(
    X_train,
    y_train
)
```

```python
print("Best parameters:")
print(grid_search.best_params_)
```

```python
print("Best cross-validation score: {:.3f}".format(
    grid_search.best_score_
))
```

### Discussion

Grid Search automates hyperparameter tuning by evaluating multiple parameter combinations.

This process helps identify settings that produce the best performance.

# Hyperparameter Tuning

Hyperparameters control the behavior of machine learning algorithms and must be selected before training.

Examples include:

- Number of neighbors in KNN
- Tree depth in Decision Trees
- Learning rate in Gradient Boosting
- C and gamma in SVM

```python
print("Best test score: {:.3f}".format(
    grid_search.score(
        X_test,
        y_test
    )
))
```

### Discussion

Proper hyperparameter tuning can significantly improve model performance and reduce overfitting.

# Confusion Matrix

A confusion matrix summarizes classification results by comparing actual labels with predicted labels.

It provides more detailed information than accuracy alone.

```python
from sklearn.metrics import confusion_matrix

svc = SVC(
    gamma=0.01,
    C=10
)

svc.fit(
    X_train,
    y_train
)

predictions = svc.predict(
    X_test
)

confusion_matrix(
    y_test,
    predictions
)
```

```python
cm = confusion_matrix(
    y_test,
    predictions
)

print(cm)
```

### Discussion

The confusion matrix reveals which classes are correctly classified and which classes are confused with one another.

This information is useful for diagnosing model weaknesses.

# Precision, Recall, and F1 Score

Accuracy alone may not provide a complete picture of model performance.

Precision, Recall, and F1 Score are commonly used to evaluate classification models, especially for imbalanced datasets.

```python
from sklearn.metrics import classification_report

print(
    classification_report(
        y_test,
        predictions
    )
)
```

### Discussion

Precision measures how many predicted positive cases are actually positive.

Recall measures how many actual positive cases are successfully detected.

The F1 Score combines precision and recall into a single metric.

# Model Comparison

Model evaluation should consider multiple metrics rather than relying on a single score.

Cross-validation, confusion matrices, and classification metrics provide complementary perspectives on model performance.

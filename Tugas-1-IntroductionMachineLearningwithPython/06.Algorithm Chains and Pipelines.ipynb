## Chapter Summary

Machine learning workflows often involve multiple processing steps before training a model.

This chapter introduces pipelines, which provide a structured way to combine preprocessing, feature selection, and model training into a single workflow.

Pipelines improve reproducibility, reduce errors, and simplify model evaluation and hyperparameter tuning.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

%matplotlib inline
```

# Pipelines

A pipeline combines multiple machine learning steps into a single object.

Typical steps include:

- Data preprocessing
- Feature selection
- Model training

Using pipelines helps ensure that all transformations are applied consistently.

```python
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data,
    cancer.target,
    random_state=0
)
```

```python
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)

svm = SVC()

svm.fit(
    X_train_scaled,
    y_train
)
```

```python
X_test_scaled = scaler.transform(X_test)

print("Test score: {:.3f}".format(
    svm.score(
        X_test_scaled,
        y_test
    )
))
```

### Discussion

Without a pipeline, each preprocessing step must be manually applied.

This increases the risk of mistakes and makes workflows harder to maintain.

# Building a Pipeline

Scikit-learn provides the Pipeline class, which allows multiple processing steps to be connected together.

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ("scaler", MinMaxScaler()),
    ("svm", SVC())
])

pipe.fit(
    X_train,
    y_train
)
```

```python
print("Pipeline score: {:.3f}".format(
    pipe.score(
        X_test,
        y_test
    )
))
```

### Discussion

The pipeline automatically applies preprocessing before model training and prediction.

This makes machine learning workflows cleaner and more reliable.

# Pipelines with Grid Search

Pipelines become especially useful when combined with hyperparameter tuning.

Grid Search can evaluate multiple preprocessing and model configurations simultaneously.

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "svm__C": [0.1, 1, 10],
    "svm__gamma": [0.001, 0.01, 0.1]
}
```

```python
grid = GridSearchCV(
    pipe,
    param_grid,
    cv=5
)

grid.fit(
    X_train,
    y_train
)
```

```python
print("Best parameters:")
print(grid.best_params_)
```

```python
print("Best cross-validation score: {:.3f}".format(
    grid.best_score_
))
```

### Discussion

Grid Search explores different parameter combinations and identifies the configuration that produces the best validation performance.

Combining Grid Search with pipelines ensures that preprocessing is correctly included in every evaluation.

# Feature Selection in Pipelines

Feature selection can also be integrated into pipelines.

This allows feature selection to be performed automatically during model training and evaluation.

```python
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import f_classif

pipe = Pipeline([
    ("select", SelectPercentile(
        f_classif,
        percentile=50
    )),
    ("svm", SVC())
])
```

```python
pipe.fit(
    X_train,
    y_train
)

print("Pipeline score: {:.3f}".format(
    pipe.score(
        X_test,
        y_test
    )
))
```

### Discussion

Feature selection inside a pipeline prevents data leakage and ensures that the evaluation process remains fair.

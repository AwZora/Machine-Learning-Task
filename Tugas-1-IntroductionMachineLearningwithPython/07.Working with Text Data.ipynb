# Bag of Words

Machine learning models cannot directly process raw text.

The Bag-of-Words representation converts text into numerical features by counting how often words appear in a document.

```python
bards_words = [
    "The fool doth think he is wise",
    "but the wise man knows himself to be a fool"
]

bards_words
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()

vect.fit(bards_words)
```

```python
print("Vocabulary size:",
      len(vect.vocabulary_))

print(vect.vocabulary_)
```

```python
bag_of_words = vect.transform(
    bards_words
)

print("Bag of words matrix:")
print(bag_of_words.toarray())
```

### Discussion

CountVectorizer converts text into a numerical matrix where each column represents a unique word and each row represents a document.

The resulting representation allows machine learning algorithms to process textual information.

# Working with Larger Text Collections

Real-world text classification tasks often involve thousands of documents.

Bag-of-Words remains one of the simplest and most effective methods for representing textual information.

```python
text_train = [
    "I love machine learning",
    "Machine learning is amazing",
    "I enjoy data science",
    "Data science is interesting"
]

vect = CountVectorizer()

X_train = vect.fit_transform(
    text_train
)

print(X_train.toarray())
```

### Discussion

Each document is transformed into a vector containing word frequencies.

Documents containing similar words often produce similar feature vectors.

# TF-IDF

Term Frequency-Inverse Document Frequency (TF-IDF) improves the Bag-of-Words representation by reducing the influence of extremely common words.

Words that appear frequently in a specific document but not across all documents receive higher importance scores.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

X_tfidf = tfidf.fit_transform(
    text_train
)

print(X_tfidf.shape)
```

```python
print(tfidf.get_feature_names_out())
```

### Discussion

TF-IDF helps highlight words that are more informative for distinguishing documents.

This often improves classification performance compared to simple word counts.

# Text Classification

Text classification is a supervised learning task where documents are assigned to predefined categories.

Examples include spam detection, sentiment analysis, and topic classification.

```python
documents = [
    "I love this movie",
    "This movie is fantastic",
    "I hate this movie",
    "This movie is terrible"
]

labels = [
    1,
    1,
    0,
    0
]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    documents,
    labels,
    random_state=42
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

text_clf = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", LogisticRegression())
])
```

```python
text_clf.fit(
    X_train,
    y_train
)
```

```python
print("Test accuracy: {:.3f}".format(
    text_clf.score(
        X_test,
        y_test
    )
))
```

### Discussion

The pipeline automatically converts text into TF-IDF features and then trains a Logistic Regression classifier.

This workflow is commonly used in text classification problems.

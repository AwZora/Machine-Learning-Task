## Theory: Eigenvectors

An eigenvector is a special vector whose direction remains unchanged after a matrix transformation.

For a matrix A and vector v:

Av = λv

where:

- v is an eigenvector
- λ is an eigenvalue

The transformation changes only the magnitude of the vector.

```python
import numpy as np

A = np.array([
    [3, 1],
    [0, 2]
])

print(A)
```

## Code Explanation

The matrix A will be used throughout this chapter to demonstrate eigenvalues and eigenvectors.

The goal is to identify vectors that maintain their direction after transformation by A.

## Theory: Eigenvalues

Eigenvalues describe how much an eigenvector is stretched or compressed during transformation.

If:

Av = λv

then λ indicates the scaling factor.

Large eigenvalues correspond to directions with greater variance or influence.

```python
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Eigenvalues:")
print(eigenvalues)
```

## Code Explanation

The eig() function computes the eigenvalues and eigenvectors of a square matrix.

The resulting eigenvalues describe the scaling behavior of the matrix.

```python
print("Eigenvectors:")
print(eigenvectors)
```

## Theory: Interpreting Eigenvectors

Each column of the eigenvector matrix corresponds to an eigenvalue.

Eigenvectors identify special directions that remain unchanged under the matrix transformation.

These directions often reveal important patterns within data.

```python
v = eigenvectors[:,0]

print(v)
```

```python
lambda_1 = eigenvalues[0]

Av = A @ v

lambda_v = lambda_1 * v

print("Av:")
print(Av)

print("\nλv:")
print(lambda_v)
```

## Code Explanation

The equality:

Av = λv

can be verified numerically.

If both results match, the vector is an eigenvector associated with the selected eigenvalue.

## Theory: Eigendecomposition Formula

A matrix can be decomposed into three matrices:

A = VDV⁻¹

where:

- A is the original matrix
- V contains eigenvectors
- D contains eigenvalues
- V⁻¹ is the inverse of V

This factorization reveals the internal structure of a matrix and simplifies many computations.

```python
V = eigenvectors

D = np.diag(eigenvalues)

V_inv = np.linalg.inv(V)

print("V:")
print(V)

print("\nD:")
print(D)

print("\nV^-1:")
print(V_inv)
```

## Code Explanation

The eigendecomposition separates the matrix into eigenvectors and eigenvalues.

This representation often makes mathematical analysis easier than working with the original matrix directly.

```python
A_reconstructed = V @ D @ V_inv

print(A_reconstructed)
```

## Theory: Matrix Reconstruction

If the eigendecomposition is correct, multiplying:

VDV⁻¹

should reconstruct the original matrix.

This provides a useful verification of the decomposition.

```python
print("Original Matrix:")
print(A)

print("\nReconstructed Matrix:")
print(A_reconstructed)
```

## Theory: Symmetric Matrices

Symmetric matrices satisfy:

A = Aᵀ

Symmetric matrices have several useful properties:

- Real eigenvalues
- Orthogonal eigenvectors
- Improved numerical stability

Many machine learning applications involve symmetric matrices.

```python
S = np.array([
    [2, 1],
    [1, 2]
])

eigvals_S, eigvecs_S = np.linalg.eig(S)

print("Eigenvalues:")
print(eigvals_S)

print("\nEigenvectors:")
print(eigvecs_S)
```

## Code Explanation

The symmetric matrix produces real eigenvalues and orthogonal eigenvectors.

These properties make symmetric matrices particularly important in machine learning and statistics.

## Theory: Dominant Eigenvalue

The dominant eigenvalue is the eigenvalue with the largest magnitude.

Its corresponding eigenvector often captures the most important direction of variation in the data.

This idea is central to Principal Component Analysis (PCA).

```python
dominant_index = np.argmax(
    np.abs(eigenvalues)
)

dominant_eigenvalue = eigenvalues[
    dominant_index
]

print(dominant_eigenvalue)
```

## Theory: Eigenvectors and Data Variance

In many datasets, directions associated with larger eigenvalues contain more information.

Dimensionality reduction techniques preserve these important directions while discarding less informative ones.

```python
dominant_eigenvector = eigenvectors[
    :,
    dominant_index
]

print(dominant_eigenvector)
```

## Theory: Principal Component Analysis

PCA uses eigenvalues and eigenvectors of a covariance matrix.

The principal components correspond to eigenvectors associated with the largest eigenvalues.

PCA is widely used for:

- Feature reduction
- Data visualization
- Noise reduction

```python
cov_matrix = np.array([
    [3, 1],
    [1, 2]
])

eigvals_cov, eigvecs_cov = np.linalg.eig(
    cov_matrix
)

print(eigvals_cov)
```

## Code Explanation

The eigenvalues indicate how much variance exists along each principal direction.

Larger eigenvalues correspond to more informative directions.

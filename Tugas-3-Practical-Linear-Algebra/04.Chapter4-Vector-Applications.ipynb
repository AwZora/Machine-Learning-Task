## Theory: Feature Vectors

In machine learning, observations are commonly represented as feature vectors.

Example:

Student A:

[Age, GPA, StudyHours]

[20, 3.8, 25]

Each element represents a measurable characteristic.

Feature vectors allow machine learning algorithms to process data numerically.

```python
import numpy as np

student = np.array([
    20,
    3.8,
    25
])

print(student)
```

## Code Explanation

The vector represents a student using numerical features.

Many machine learning datasets store observations in this format because vectors provide a compact and structured representation.

## Theory: Similarity Measurement

One of the most important applications of vectors is measuring similarity.

Similarity measurement is widely used in:

- Recommendation Systems
- Search Engines
- Text Analysis
- Customer Segmentation

Cosine Similarity is one of the most popular similarity metrics.

```python
a = np.array([1, 2, 3])

b = np.array([2, 4, 6])

cosine_similarity = np.dot(a, b) / (
    np.linalg.norm(a)
    *
    np.linalg.norm(b)
)

print(cosine_similarity)
```

## Code Explanation

The cosine similarity measures the angle between vectors.

Values close to:

- 1 indicate strong similarity.
- 0 indicate independence.
- -1 indicate opposite directions.

Cosine similarity is widely used in recommendation systems and natural language processing.

## Theory: Distance Measurement

Distance measures how far two observations are from each other.

Common distance metrics include:

- Euclidean Distance
- Manhattan Distance
- Cosine Distance

Distance metrics play a central role in clustering and nearest-neighbor algorithms.

```python
point1 = np.array([1, 2])

point2 = np.array([4, 6])

distance = np.linalg.norm(
    point1 - point2
)

print(distance)
```

## Theory: Vector Transformations

Vector transformations modify vectors while preserving mathematical structure.

Examples:

- Rotation
- Scaling
- Reflection

Transformations are important in computer graphics, robotics, and machine learning.

```python
vector = np.array([2, 3])

scaled_vector = 3 * vector

print(scaled_vector)
```

## Code Explanation

Scaling multiplies every vector element by a constant.

The direction remains unchanged while the magnitude changes.

Scaling is commonly used during data preprocessing.

```python
import matplotlib.pyplot as plt

v = np.array([2, 3])

plt.figure(figsize=(6,6))

plt.quiver(
    0,0,
    v[0],v[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(0,5)
plt.ylim(0,5)

plt.grid()

plt.title("Vector Representation")

plt.show()
```

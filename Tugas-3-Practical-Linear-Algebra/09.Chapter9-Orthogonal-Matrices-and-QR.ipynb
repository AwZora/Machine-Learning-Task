## Theory: Orthogonal Vectors

Two vectors are orthogonal when their dot product equals zero.

Example:

v₁ · v₂ = 0

Orthogonality indicates that vectors are perpendicular to each other.

Orthogonal vectors play an important role in geometry, statistics, and machine learning.

```python
import numpy as np

v1 = np.array([1, 0])

v2 = np.array([0, 1])

print(np.dot(v1, v2))
```

## Code Explanation

The dot product equals zero.

This confirms that the vectors are orthogonal.

Orthogonal vectors are independent and share no directional overlap.

## Theory: Orthonormal Vectors

Orthonormal vectors satisfy two conditions:

1. Orthogonal to each other.
2. Magnitude equals one.

Orthonormal bases simplify many matrix operations and decompositions.

```python
u1 = np.array([1, 0])

u2 = np.array([0, 1])

print(np.linalg.norm(u1))
print(np.linalg.norm(u2))
```

## Code Explanation

Both vectors have magnitude equal to one.

Since they are also orthogonal, they form an orthonormal basis.

## Theory: Orthogonal Matrices

A matrix is orthogonal if:

QᵀQ = I

where:

- Qᵀ is the transpose.
- I is the identity matrix.

Orthogonal matrices preserve distances and angles.

```python
Q = np.array([
    [0, -1],
    [1,  0]
])

print(Q.T @ Q)
```

## Theory: Properties of Orthogonal Matrices

Important properties:

- Q⁻¹ = Qᵀ
- Preserve vector lengths.
- Preserve angles.
- Numerically stable.

These properties make orthogonal matrices valuable in scientific computing.

```python
Q_inv = np.linalg.inv(Q)

print(Q_inv)
print("\nTranspose:")
print(Q.T)
```

## Theory: QR Decomposition

QR decomposition factors a matrix into:

A = QR

Where:

- Q = Orthogonal Matrix
- R = Upper Triangular Matrix

QR decomposition is widely used in numerical linear algebra and least squares problems.

```python
A = np.array([
    [1, 2],
    [3, 4]
])

Q, R = np.linalg.qr(A)

print("Q Matrix:")
print(Q)

print("\nR Matrix:")
print(R)
```

## Code Explanation

The QR decomposition separates a matrix into an orthogonal component and an upper triangular component.

This decomposition simplifies many numerical computations.

```python
reconstructed = Q @ R

print(reconstructed)
```

## Theory: Dot Product

The dot product is one of the most important vector operations.

For vectors:

a = [a₁,a₂,...]

b = [b₁,b₂,...]

The dot product is calculated by multiplying corresponding elements and summing the results.

The dot product produces a scalar value.

Applications include:

- Similarity measurement
- Projections
- Machine learning optimization

```python
import numpy as np

a = np.array([1, 2, 3])

b = np.array([4, 5, 6])

dot_product = np.dot(a, b)

print(dot_product)
```

## Code Explanation

The dot product multiplies corresponding elements and sums the results.

Example:

(1×4)+(2×5)+(3×6)

= 32

The result is a scalar rather than a vector.

## Theory: Geometric Interpretation of Dot Product

The dot product also describes the relationship between two vectors geometrically.

Properties:

- Large positive value → Similar direction
- Zero → Orthogonal vectors
- Negative value → Opposite direction

This interpretation is widely used in machine learning and data analysis.

```python
v1 = np.array([1, 0])

v2 = np.array([0, 1])

print(np.dot(v1, v2))
```

## Code Explanation

The dot product equals zero.

This indicates that the vectors are perpendicular to each other.

Perpendicular vectors are called orthogonal vectors.

## Theory: Orthogonal Vectors

Two vectors are orthogonal when they form a 90-degree angle.

Orthogonal vectors satisfy:

Dot Product = 0

Orthogonality plays a major role in:

- PCA
- QR Decomposition
- Signal Processing
- Data Compression

```python
v1 = np.array([2, 1])

v2 = np.array([-1, 2])

print(np.dot(v1, v2))
```

## Theory: Vector Angle

The angle between vectors provides information about similarity.

Smaller angles indicate greater similarity.

Larger angles indicate less similarity.

This concept forms the basis of cosine similarity used in recommendation systems and text analysis.

```python
a = np.array([1, 2])

b = np.array([3, 1])

cos_theta = np.dot(a, b) / (
    np.linalg.norm(a)
    *
    np.linalg.norm(b)
)

print(cos_theta)
```

```python
angle = np.arccos(cos_theta)

print("Angle (Radians):")
print(angle)

print("\nAngle (Degrees):")
print(np.degrees(angle))
```

## Code Explanation

The angle is computed using the dot product formula.

The cosine value is converted into an angle using the inverse cosine function.

Angles provide a quantitative measure of vector similarity.

## Theory: Vector Projection

Projection measures how much one vector extends in the direction of another vector.

Vector projection is important for:

- Dimensionality reduction
- Optimization
- Signal decomposition
- Machine learning algorithms

Projection extracts useful information from high-dimensional data.

```python
a = np.array([3, 4])

b = np.array([1, 0])

projection = (
    np.dot(a, b)
    /
    np.dot(b, b)
) * b

print(projection)
```

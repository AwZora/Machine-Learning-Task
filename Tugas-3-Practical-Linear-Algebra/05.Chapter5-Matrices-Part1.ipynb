## Theory: What is a Matrix?

A matrix is a rectangular arrangement of numbers organized into rows and columns.

Example:

[[1,2,3],
 [4,5,6]]

Matrices are used to represent datasets, images, transformations, and machine learning features.

They are fundamental building blocks in modern data science.

```python
import numpy as np

A = np.array([
    [1,2,3],
    [4,5,6]
])

print(A)
```

## Code Explanation

The matrix contains two rows and three columns.

Each entry is identified by its row and column position.

Matrices provide a structured representation of numerical information.

## Theory: Matrix Dimensions

The size of a matrix is described by:

Rows × Columns

Examples:

2 × 3

3 × 3

4 × 2

Matrix dimensions determine which operations are mathematically valid.

```python
print(A.shape)
```

## Code Explanation

The shape function returns the dimensions of a matrix.

The output:

(2,3)

means:

- 2 rows
- 3 columns

## Theory: Matrix Elements

Every matrix entry has a specific location.

Notation:

A(i,j)

where:

- i = row
- j = column

Accessing individual elements is important for data manipulation and numerical computation.

```python
print(A[0,0])
print(A[1,2])
```

## Theory: Matrix Transpose

The transpose of a matrix swaps rows and columns.

Example:

Original:

[[1,2,3],
 [4,5,6]]

Transpose:

[[1,4],
 [2,5],
 [3,6]]

Transpose operations are widely used in machine learning and matrix multiplication.

```python
AT = A.T

print(AT)
```

## Code Explanation

The transpose operation changes matrix orientation.

Rows become columns and columns become rows.

This operation appears frequently in regression, optimization, and neural networks.

## Theory: Matrix Addition

Matrices can be added only when they have identical dimensions.

Addition is performed element by element.

Example:

[[1,2],
 [3,4]]

+

[[5,6],
 [7,8]]

```python
B = np.array([
    [1,1,1],
    [1,1,1]
])

C = A + B

print(C)
```

## Theory: Matrix Subtraction

Matrix subtraction follows the same rules as matrix addition.

Corresponding elements are subtracted from each other.

Matrices must have matching dimensions.

```python
D = A - B

print(D)
```

## Key Takeaways

- Matrices organize numerical information.
- Matrix dimensions describe size.
- Matrix elements can be accessed individually.
- Transpose swaps rows and columns.
- Matrix addition and subtraction require matching dimensions.

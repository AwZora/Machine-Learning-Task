## Theory: Matrix Multiplication

Matrix multiplication combines information from two matrices.

Unlike addition and subtraction, multiplication requires compatible dimensions.

Rule:

(m × n) × (n × p)

Result:

(m × p)

Matrix multiplication is one of the most important operations in machine learning.

```python
import numpy as np

A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

C = A @ B

print(C)
```

## Code Explanation

The @ operator performs matrix multiplication.

Each element in the resulting matrix is computed by multiplying rows from the first matrix with columns from the second matrix.

Matrix multiplication differs from element-wise multiplication.

## Theory: Element-Wise Multiplication

Element-wise multiplication multiplies corresponding elements directly.

This operation requires matrices of identical dimensions.

Element-wise multiplication is different from standard matrix multiplication.

```python
elementwise = A * B

print(elementwise)
```

## Code Explanation

Each element of A is multiplied by the corresponding element of B.

This operation is commonly used in numerical computing and neural networks.

## Theory: Identity Matrix

An identity matrix is a square matrix with:

- Ones on the main diagonal
- Zeros elsewhere

Example:

[[1,0],
 [0,1]]

Identity matrices behave similarly to the number 1 in scalar multiplication.

```python
I = np.eye(3)

print(I)
```

## Theory: Matrix Identity Property

For any compatible matrix A:

A × I = A

I × A = A

The identity matrix leaves the original matrix unchanged.

```python
A = np.array([
    [1,2],
    [3,4]
])

I = np.eye(2)

print(A @ I)
```

## Theory: Diagonal Matrix

A diagonal matrix contains values only on the main diagonal.

All off-diagonal entries are zero.

Diagonal matrices simplify many matrix computations and appear frequently in machine learning.

```python
D = np.diag([1,2,3])

print(D)
```

## Theory: Symmetric Matrix

A matrix is symmetric when:

A = Aᵀ

This means the matrix equals its transpose.

Symmetric matrices play a major role in:

- PCA
- Covariance matrices
- Eigendecomposition

```python
S = np.array([
    [1,2,3],
    [2,4,5],
    [3,5,6]
])

print(np.array_equal(S, S.T))
```

## Theory: What is a Matrix Inverse?

The inverse of a matrix is similar to division in arithmetic.

For a matrix A:

A⁻¹

is called the inverse matrix.

The inverse satisfies:

A × A⁻¹ = I

where I is the identity matrix.

```python
import numpy as np

A = np.array([
    [1,2],
    [3,4]
])

A_inv = np.linalg.inv(A)

print(A_inv)
```

## Code Explanation

The NumPy function inv() computes the inverse of a matrix.

Only square matrices that satisfy certain conditions have inverses.

## Theory: Identity Property

If a matrix inverse exists:

A × A⁻¹ = I

where I is the identity matrix.

This property is commonly used to verify inverse calculations.

```python
result = A @ A_inv

print(result)
```

## Code Explanation

The result should be close to the identity matrix.

Small numerical differences may occur due to floating-point precision.

## Theory: Conditions for Invertibility

Not every matrix has an inverse.

Requirements:

- Matrix must be square.
- Determinant must not equal zero.

Matrices that have inverses are called invertible matrices.

Matrices without inverses are called singular matrices.

```python
determinant = np.linalg.det(A)

print(determinant)
```

## Theory: Determinants

The determinant is a numerical value associated with a square matrix.

The determinant helps determine:

- Invertibility
- Matrix scaling effects
- Geometric properties

If determinant = 0, the matrix is singular.

```python
B = np.array([
    [1,2],
    [2,4]
])

print(np.linalg.det(B))
```

## Code Explanation

The determinant is zero.

Therefore, the matrix is singular and does not have an inverse.

## Theory: Solving Linear Systems

Matrix inverses can solve systems of linear equations.

General form:

Ax = b

Solution:

x = A⁻¹b

This approach is fundamental in many machine learning algorithms.

```python
A = np.array([
    [2,1],
    [1,3]
])

b = np.array([
    [5],
    [6]
])

x = np.linalg.inv(A) @ b

print(x)
```

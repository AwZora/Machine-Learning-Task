## Theory: Scalars, Vectors, and Matrices

Linear algebra begins with three fundamental objects:

### Scalar

A single numerical value.

Example:

5

### Vector

An ordered collection of values.

Example:

[1, 2, 3]

### Matrix

A rectangular arrangement of numbers.

Example:

[[1,2],
 [3,4]]

```python
scalar = 10

vector = np.array([1, 2, 3])

matrix = np.array([
    [1, 2],
    [3, 4]
])

print("Scalar:")
print(scalar)

print("\nVector:")
print(vector)

print("\nMatrix:")
print(matrix)
```

## Theory: Python for Linear Algebra

```python
import numpy as np

a = np.array([1, 2, 3])

b = np.array([4, 5, 6])

print(a)
print(b)
```

```python
vector_sum = a + b

print("Vector Addition:")
print(vector_sum)
```

## Code Explanation

Vectors can be added element by element.

For example:

[1,2,3] + [4,5,6]

results in:

[5,7,9]

Vector addition is one of the most common operations in linear algebra.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

plt.quiver(
    0, 0,
    3, 2,
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(0,5)
plt.ylim(0,5)

plt.grid()

plt.title("Vector Visualization")

plt.show()
```

## Theory: Visualization of Vectors

## Theory: Mathematical Notation

Linear algebra uses notation to represent mathematical objects efficiently.

Common notation:

- Scalars: lowercase letters (a, b, c)
- Vectors: bold lowercase letters (**v**)
- Matrices: bold uppercase letters (**A**)

This notation helps distinguish different mathematical objects and improves readability.

```python
a = 5

v = np.array([1, 2, 3])

A = np.array([
    [1, 2],
    [3, 4]
])

print("Scalar:")
print(a)

print("\nVector:")
print(v)

print("\nMatrix:")
print(A)
```

## Code Explanation

The example demonstrates the three primary mathematical objects used throughout linear algebra.

Understanding the differences between scalars, vectors, and matrices is essential because each object follows different mathematical rules and operations.

## Theory: Dimensions

Dimensions describe the size of vectors and matrices.

Examples:

Vector:

[1,2,3]

Dimension:

3 × 1

Matrix:

[[1,2],
 [3,4]]

Dimension:

2 × 2

Understanding dimensions is important because many operations require compatible dimensions.

```python
vector = np.array([1, 2, 3])

matrix = np.array([
    [1, 2],
    [3, 4]
])

print("Vector Shape:")
print(vector.shape)

print("\nMatrix Shape:")
print(matrix.shape)
```

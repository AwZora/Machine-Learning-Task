## Theory: What is a Vector?

A vector is an ordered collection of numbers.

Unlike scalars, vectors can contain multiple values.

Examples:

[1, 2, 3]

[5, 10, 15]

Vectors are used extensively in machine learning to represent:

- Features
- Observations
- Model parameters
- Predictions

```python
import numpy as np

v = np.array([1, 2, 3])

print(v)
```

## Code Explanation

The code creates a vector containing three elements.

NumPy arrays are commonly used to represent vectors in Python because they provide efficient numerical computation.

## Theory: Vector Orientation

Vectors can be represented in two orientations:

### Row Vector

[1 2 3]

### Column Vector

[1
 2
 3]

Orientation becomes important in matrix multiplication and advanced linear algebra operations.

```python
row_vector = np.array([1, 2, 3])

column_vector = np.array([
    [1],
    [2],
    [3]
])

print("Row Vector:")
print(row_vector)

print("\nColumn Vector:")
print(column_vector)
```

## Code Explanation

The first vector is stored as a row vector.

The second vector is stored as a column vector.

Although both contain the same values, their dimensions differ.

## Theory: Vector Dimensions

The dimension of a vector corresponds to the number of elements it contains.

Examples:

[1,2] → 2-dimensional

[1,2,3] → 3-dimensional

[1,2,3,4,5] → 5-dimensional

Dimension is an important property because many operations require vectors of equal dimensions.

```python
v1 = np.array([1, 2])

v2 = np.array([1, 2, 3])

v3 = np.array([1, 2, 3, 4, 5])

print(v1.shape)
print(v2.shape)
print(v3.shape)
```

## Code Explanation

The first example generates a random vector.

The second example generates a sequential vector.

Automatically generated vectors are useful when creating synthetic datasets or performing mathematical experiments.

## Theory: Vector Addition

Vector addition combines two vectors by adding corresponding elements.

Example:

[1,2,3] + [4,5,6]

=

[5,7,9]

Vector addition requires both vectors to have the same dimensions.

```python
v1 = np.array([1, 2, 3])

v2 = np.array([4, 5, 6])

v_sum = v1 + v2

print(v_sum)
```

## Code Explanation

Each element in the first vector is added to the corresponding element in the second vector.

The result is a new vector with the same dimension.

## Theory: Vector Subtraction

Vector subtraction works similarly to vector addition.

Corresponding elements are subtracted from one another.

Example:

[5,7,9] - [1,2,3]

=

[4,5,6]

```python
v_sub = v2 - v1

print(v_sub)
```

## Theory: Scalar Multiplication

A scalar can multiply every element in a vector.

Example:

2 × [1,2,3]

=

[2,4,6]

Scalar multiplication changes the magnitude of a vector while preserving its direction.

```python
v = np.array([1, 2, 3])

scaled_v = 2 * v

print(scaled_v)
```

## Code Explanation

Every element in the vector is multiplied by the scalar value.

Scalar multiplication is widely used in machine learning optimization algorithms.

## Theory: Vector Magnitude

Vector magnitude measures the length of a vector.

For a vector:

v = [x,y]

Magnitude:

√(x² + y²)

Magnitude provides information about the size of a vector.

```python
v = np.array([3, 4])

magnitude = np.linalg.norm(v)

print(magnitude)
```

## Theory: Unit Vectors

A unit vector is a vector with magnitude equal to 1.

Unit vectors are often used to represent direction independently of magnitude.

They are important in geometry, physics, and machine learning.

```python
v = np.array([3, 4])

unit_vector = v / np.linalg.norm(v)

print(unit_vector)
```

## Code Explanation

The vector is divided by its magnitude.

The resulting vector maintains the original direction but has a magnitude of exactly one.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))

plt.quiver(
    0,0,
    unit_vector[0],
    unit_vector[1],
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(0,1)
plt.ylim(0,1)

plt.grid()

plt.title("Unit Vector")

plt.show()
```

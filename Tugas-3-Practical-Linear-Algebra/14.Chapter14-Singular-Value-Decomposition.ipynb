## Theory: What is Singular Value Decomposition?

Singular Value Decomposition factorizes a matrix A into:

A = UΣVᵀ

where:

- U contains left singular vectors.
- Σ contains singular values.
- Vᵀ contains right singular vectors.

Unlike eigendecomposition, SVD can be applied to any matrix, including non-square matrices.

```python
import numpy as np

A = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

print(A)
```

## Code Explanation

The matrix A is a rectangular matrix with three rows and two columns.

Unlike eigendecomposition, SVD can be performed on rectangular matrices.

## Theory: Components of SVD

The decomposition produces three matrices:

U → Left singular vectors

Σ → Singular values

Vᵀ → Right singular vectors

Together these matrices completely describe the original matrix.

```python
U, s, VT = np.linalg.svd(A)

print("U Matrix:")
print(U)
```

```python
print("Singular Values:")
print(s)
```

```python
print("VT Matrix:")
print(VT)
```

## Code Explanation

The svd() function computes the singular value decomposition of a matrix.

The output includes singular vectors and singular values that describe the structure of the matrix.

## Theory: Singular Values

Singular values indicate the importance of each component of the matrix.

Large singular values correspond to directions containing more information.

Small singular values often correspond to noise or less important information.

```python
Sigma = np.zeros(
    (
        A.shape[0],
        A.shape[1]
    )
)

np.fill_diagonal(
    Sigma,
    s
)

print(Sigma)
```

## Theory: Sigma Matrix

The Sigma matrix contains singular values arranged along its diagonal.

All off-diagonal entries are zero.

The singular values are typically ordered from largest to smallest.

```python
print("Rank:")
print(np.linalg.matrix_rank(A))
```

## Theory: Matrix Rank

The rank of a matrix indicates the number of independent directions contained within the data.

SVD provides a natural way to understand matrix rank through singular values.

```python
nonzero_sv = np.sum(
    s > 1e-10
)

print(nonzero_sv)
```

## Code Explanation

The number of nonzero singular values equals the rank of the matrix.

This property makes SVD useful for analyzing data structure and dimensionality.

## Theory: Matrix Reconstruction

One of the most important properties of SVD is that the original matrix can be reconstructed from its decomposition.

The reconstruction formula is:

A = UΣVᵀ

If the decomposition is correct, the reconstructed matrix should closely match the original matrix.

```python
explained_information = (
    s**2 /
    np.sum(s**2)
)

print(explained_information)
```

## Theory: Information Content

The squared singular values indicate how much information is captured by each component.

Components associated with larger singular values explain more variance in the data.

```python
cumulative_information = np.cumsum(
    explained_information
)

print(cumulative_information)
```

## Theory: Relationship Between SVD and PCA

Principal Component Analysis (PCA) is closely related to SVD.

PCA identifies directions of maximum variance in data.

In practice, many PCA implementations use SVD because it is numerically stable and computationally efficient.

```python
X = np.array([
    [2,1],
    [3,2],
    [4,3],
    [5,4]
])

X_centered = X - np.mean(
    X,
    axis=0
)

U_pca, s_pca, VT_pca = np.linalg.svd(
    X_centered
)

print(VT_pca)
```

## Code Explanation

The right singular vectors identify the principal directions within the dataset.

These directions correspond to principal components in PCA.

## Theory: Image Compression

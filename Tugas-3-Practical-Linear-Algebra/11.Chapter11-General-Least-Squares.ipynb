## Theory: Overdetermined Systems

An overdetermined system contains more equations than unknown variables.

Example:

x + y = 2

2x + y = 3

3x + y = 5

In many real-world situations, no exact solution satisfies all equations simultaneously.

Least Squares provides the best approximate solution.

```python
import numpy as np

A = np.array([
    [1,1],
    [2,1],
    [3,1]
])

b = np.array([
    [2],
    [3],
    [5]
])

print(A)
print()
print(b)
```

## Code Explanation

The matrix contains three equations and two unknown variables.

Because there are more equations than variables, the system is overdetermined.

## Theory: Exact vs Approximate Solutions

For many datasets, measurement errors and noise prevent exact solutions.

Instead of finding a perfect solution, Least Squares finds the solution that minimizes overall error.

This makes it extremely useful in machine learning and statistics.

```python
x, residuals, rank, s = np.linalg.lstsq(
    A,
    b,
    rcond=None
)

print("Solution:")
print(x)
```

## Code Explanation

The lstsq() function computes the least squares solution.

Instead of satisfying every equation exactly, it minimizes the total prediction error.

## Theory: Residuals

Residuals measure the difference between observed values and predicted values.

Residual:

Observed - Predicted

Residuals indicate how well a model fits the data.

```python
predictions = A @ x

residuals_vector = b - predictions

print(residuals_vector)
```

## Theory: Error Minimization

Least Squares minimizes the sum of squared residuals.

Objective:

Minimize:

Σ(residual²)

Squaring prevents positive and negative errors from canceling each other.

```python
sse = np.sum(
    residuals_vector**2
)

print("Sum of Squared Errors:")
print(sse)
```

## Code Explanation

The Sum of Squared Errors (SSE) quantifies the total prediction error.

Smaller SSE values indicate better model fit.

## Theory: Geometric Interpretation

Least Squares can be interpreted geometrically.

The solution corresponds to projecting the target vector onto the column space of the matrix.

This projection minimizes the distance between observations and predictions.

```python
print("Predictions:")
print(predictions)
```

## Theory: Column Space

The column space of a matrix contains all possible linear combinations of its columns.

Least Squares finds the projection of the target vector onto this space.

This interpretation connects linear algebra with machine learning optimization.

```python
rank = np.linalg.matrix_rank(A)

print("Rank:")
print(rank)
```

## Theory: Normal Equation

The Least Squares solution can be derived mathematically using the Normal Equation.

Given:

Ax = b

The Least Squares solution is:

x = (AᵀA)⁻¹Aᵀb

This equation provides the vector that minimizes the sum of squared residuals.

```python
A = np.array([
    [1,1],
    [2,1],
    [3,1]
])

b = np.array([
    [2],
    [3],
    [5]
])

x_normal = np.linalg.inv(
    A.T @ A
) @ A.T @ b

print(x_normal)
```

## Code Explanation

The Normal Equation computes the Least Squares solution directly using matrix operations.

The resulting vector contains the coefficients that minimize prediction error.

## Theory: Why the Normal Equation Works

The Normal Equation is derived by minimizing the Sum of Squared Errors (SSE).

By setting the derivative of the error function equal to zero, the optimal solution can be obtained analytically.

This approach forms the mathematical foundation of Linear Regression.

```python
predictions = A @ x_normal

print(predictions)
```

```python
residuals = b - predictions

print(residuals)
```

## Code Explanation

Residuals measure the difference between actual values and predicted values.

A good Least Squares solution minimizes these residuals.

## Theory: Sum of Squared Errors

The objective of Least Squares is to minimize:

SSE = Σ(residual²)

Squaring emphasizes larger errors and ensures all errors contribute positively to the objective function.

```python
sse = np.sum(
    residuals**2
)

print(sse)
```

## Theory: Projection Interpretation

The Least Squares solution can be interpreted geometrically as a projection.

The prediction vector is the projection of the target vector onto the column space of the matrix.

This projection produces the smallest possible error.

```python
projection = A @ x_normal

print(projection)
```

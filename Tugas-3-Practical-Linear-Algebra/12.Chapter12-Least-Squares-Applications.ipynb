## Theory: Model Fitting

Model fitting refers to the process of finding mathematical relationships within data.

Least Squares determines model parameters that minimize prediction errors.

This process forms the basis of supervised machine learning.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])

y = np.array([2,4,5,4,5])

plt.scatter(x,y)

plt.title("Observed Data")

plt.show()
```

## Code Explanation

The scatter plot represents observed data points.

The objective of model fitting is to identify a mathematical relationship that explains the observed pattern.

## Theory: Linear Regression

Linear Regression assumes a linear relationship between variables.

General form:

y = mx + b

where:

- m = slope
- b = intercept

The values of m and b are estimated using Least Squares.

```python
coefficients = np.polyfit(
    x,
    y,
    1
)

print(coefficients)
```

## Code Explanation

The polyfit() function estimates the slope and intercept of the best-fitting line.

These values are chosen to minimize the sum of squared errors.

```python
slope = coefficients[0]

intercept = coefficients[1]

predictions = slope * x + intercept

print(predictions)
```

## Theory: Predictions

Once model parameters are estimated, predictions can be generated.

Predictions represent the model's estimate of the target variable.

Machine learning models use this process to make forecasts and classifications.

```python
plt.scatter(x,y)

plt.plot(
    x,
    predictions
)

plt.title("Linear Regression Fit")

plt.show()
```

## Code Explanation

The line represents the Least Squares solution.

The model attempts to pass as close as possible to all observed points while minimizing prediction error.

## Theory: Model Error

No model perfectly describes real-world data.

Error measures how far predictions differ from observations.

Evaluating error is essential for assessing model quality.

```python
errors = y - predictions

print(errors)
```

```python
sse = np.sum(
    errors**2
)

print("SSE:")
print(sse)
```

## Code Explanation

The Sum of Squared Errors (SSE) summarizes overall prediction error.

Smaller SSE values indicate a better fit.

## Theory: Trend Analysis

Least Squares is commonly used to identify trends within data.

Trend analysis supports:

- Forecasting
- Financial modeling
- Sales prediction
- Scientific analysis

Trend detection is one of the most common uses of regression models.

```python
future_x = np.array([6,7,8])

future_predictions = (
    slope * future_x
    + intercept
)

print(future_predictions)
```

## Theory: Multiple Linear Regression

Many real-world problems involve multiple input variables.

Multiple Linear Regression extends simple linear regression by using several predictors simultaneously.

General form:

y = β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ

Least Squares is used to estimate the coefficients that best fit the data.

```python
X = np.array([
    [1, 2],
    [2, 1],
    [3, 4],
    [4, 3]
])

y = np.array([
    [5],
    [4],
    [9],
    [8]
])

print(X)
print(y)
```

## Code Explanation

The matrix X contains two predictor variables.

Each row represents an observation, while each column represents a feature.

The vector y contains the target values.

## Theory: Matrix Representation of Regression

Multiple regression can be written in matrix form:

y = Xβ

where:

- X is the feature matrix.
- β contains model coefficients.
- y is the target vector.

This representation allows efficient computation using linear algebra.

```python
X_design = np.column_stack(
    (
        np.ones(X.shape[0]),
        X
    )
)

print(X_design)
```

## Code Explanation

A column of ones is added to represent the intercept term.

This augmented matrix is commonly called the design matrix.

```python
beta = np.linalg.inv(
    X_design.T @ X_design
) @ X_design.T @ y

print(beta)
```

## Theory: Estimating Regression Coefficients

The Normal Equation provides the Least Squares solution for multiple regression.

The resulting coefficients describe the contribution of each predictor variable to the target variable.

```python
predictions = X_design @ beta

print(predictions)
```

## Theory: Residual Analysis

Residuals measure the difference between actual and predicted values.

Analyzing residuals helps evaluate model quality and identify potential modeling issues.

```python
residuals = y - predictions

print(residuals)
```

```python
sse = np.sum(
    residuals**2
)

print("SSE:")
print(sse)
```

## Code Explanation

The Sum of Squared Errors quantifies overall prediction error.

Lower SSE values generally indicate a better model fit.

## Theory: Model Evaluation

Evaluating a model is an important step in machine learning.

Common evaluation metrics include:

- Mean Squared Error (MSE)
- Root Mean Squared Error (RMSE)
- R² Score

These metrics help determine how well the model explains the data.

```python
mse = np.mean(
    residuals**2
)

rmse = np.sqrt(mse)

print("MSE:", mse)
print("RMSE:", rmse)
```

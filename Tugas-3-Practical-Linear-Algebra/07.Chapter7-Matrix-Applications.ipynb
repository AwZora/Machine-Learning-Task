## Theory: Data as Matrices

Most datasets can be represented as matrices.

Example:

| Age | Salary | Experience |
|------|---------|------------|
| 25 | 5000 | 2 |
| 30 | 7000 | 5 |
| 35 | 9000 | 8 |

Rows represent observations.

Columns represent features.

This structure is commonly called a data matrix.

```python
import numpy as np

data = np.array([
    [25, 5000, 2],
    [30, 7000, 5],
    [35, 9000, 8]
])

print(data)
```

## Code Explanation

The matrix stores multiple observations and features.

This format is used by most machine learning libraries including Scikit-Learn, TensorFlow, and PyTorch.

## Theory: Matrix Transformations

Matrix transformations modify data while preserving mathematical structure.

Examples include:

- Scaling
- Rotation
- Reflection
- Projection

Transformations are widely used in computer graphics and machine learning.

```python
A = np.array([
    [1, 2],
    [3, 4]
])

scaled_A = 2 * A

print(scaled_A)
```

## Code Explanation

The matrix is scaled by multiplying every element by a constant value.

Scaling is frequently used during data preprocessing and normalization.

## Theory: Images as Matrices

Digital images can be represented as matrices.

For grayscale images:

- Each pixel corresponds to a numerical value.
- Larger values represent brighter pixels.

Image processing algorithms operate directly on these matrix representations.

```python
image_matrix = np.array([
    [0, 50, 100],
    [150, 200, 255],
    [255, 200, 100]
])

print(image_matrix)
```

## Theory: Matrix Operations in Machine Learning

Machine learning models often use matrices to:

- Store datasets
- Store model parameters
- Generate predictions
- Compute gradients

Matrix operations make these computations efficient even for large datasets.

```python
X = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

weights = np.array([
    [0.5],
    [1.0]
])

predictions = X @ weights

print(predictions)
```

## Code Explanation

The matrix multiplication combines feature values with model weights.

This operation is the foundation of linear regression, logistic regression, and neural networks.

## Theory: Covariance Matrices

Covariance matrices describe relationships among variables.

Applications include:

- Principal Component Analysis (PCA)
- Feature Analysis
- Statistical Modeling

Covariance matrices are usually symmetric matrices.

```python
data = np.array([
    [1,2],
    [3,4],
    [5,6]
])

cov_matrix = np.cov(
    data.T
)

print(cov_matrix)
```

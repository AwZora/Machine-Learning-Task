## Theory: k-Nearest Neighbors (k-NN)

k-Nearest Neighbors is a simple machine learning algorithm that classifies observations based on the nearest training examples.

Procedure:

1. Choose a value of k.
2. Calculate distances.
3. Identify nearest neighbors.
4. Assign the majority class.

Advantages:

- Easy to understand.
- No training phase.

Disadvantages:

- Computationally expensive for large datasets.
- Sensitive to irrelevant features.

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd

data = pd.DataFrame({
    'Age': [20,25,30,35,40,45,50,55],
    'Purchased': [0,0,0,1,1,1,1,1]
})

X = data[['Age']]
y = data['Purchased']

knn = KNeighborsClassifier(
    n_neighbors=3
)

knn.fit(X, y)

predictions = knn.predict(X)

print(predictions)
```

## Code Explanation

The k-NN algorithm predicts class labels based on the nearest observations.

The parameter k determines the number of neighbors considered during prediction.

Smaller values of k create more flexible models, while larger values create smoother decision boundaries.

## Theory: Decision Trees

Decision Trees divide data into smaller subsets using a sequence of decision rules.

Each node represents a decision point.

Each branch represents a possible outcome.

Each leaf node represents a final prediction.

Decision Trees are intuitive and easy to interpret.

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier(
    random_state=42
)

tree_model.fit(X, y)

tree_predictions = tree_model.predict(X)

print(tree_predictions)
```

```python
from sklearn.metrics import accuracy_score

tree_accuracy = accuracy_score(
    y,
    tree_predictions
)

print("Decision Tree Accuracy:")
print(tree_accuracy)
```

## Discussion

k-Nearest Neighbors and Decision Trees are fundamental machine learning algorithms.

k-NN classifies observations based on similarity, while Decision Trees use hierarchical decision rules.

Both methods provide interpretable solutions and are frequently used as baseline machine learning models.

## Theory: Ensemble Learning

Ensemble Learning combines multiple machine learning models to improve predictive performance.

The main idea is that multiple models working together often perform better than a single model.

Benefits:

- Improved accuracy
- Reduced overfitting
- Increased robustness

Popular ensemble methods include:

- Bagging
- Random Forest
- Boosting

## Theory: Bagging

Bagging stands for Bootstrap Aggregating.

The process:

1. Generate multiple bootstrap samples.
2. Train a separate model on each sample.
3. Combine predictions.

Bagging reduces variance and improves model stability.

Decision Trees are commonly used as base learners in bagging methods.

```python
from sklearn.ensemble import BaggingClassifier

bagging_model = BaggingClassifier(
    n_estimators=10,
    random_state=42
)

bagging_model.fit(X, y)

bagging_predictions = bagging_model.predict(X)

print(bagging_predictions)
```

```python
bagging_accuracy = accuracy_score(
    y,
    bagging_predictions
)

print("Bagging Accuracy:")
print(bagging_accuracy)
```

## Theory: Random Forest

Random Forest is an ensemble learning method based on multiple decision trees.

Each tree:

- Uses a bootstrap sample.
- Considers a random subset of features.

The final prediction is determined through majority voting.

Advantages:

- High predictive accuracy.
- Reduced overfitting.
- Handles large datasets effectively.

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X, y)

rf_predictions = rf_model.predict(X)

print(rf_predictions)
```

```python
rf_accuracy = accuracy_score(
    y,
    rf_predictions
)

print("Random Forest Accuracy:")
print(rf_accuracy)
```

```python
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
})

print(feature_importance)
```

## Code Explanation

Random Forest provides feature importance scores that indicate how much each feature contributes to prediction performance.

Features with higher importance values have a greater influence on model decisions.

Feature importance is useful for:

- Feature selection
- Model interpretation
- Understanding data relationships

## Discussion

Ensemble methods improve predictive performance by combining multiple models.

Bagging reduces model variance through bootstrap aggregation, while Random Forest introduces additional randomness to improve generalization.

These methods are widely used because they often outperform individual models while remaining relatively easy to implement.

## Theory: Boosting

Boosting is an ensemble learning technique that combines multiple weak learners to create a stronger predictive model.

Unlike Bagging, Boosting trains models sequentially.

Each new model focuses on correcting errors made by previous models.

Advantages:

- High predictive accuracy
- Effective for complex datasets
- Strong performance in machine learning competitions

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_model = GradientBoostingClassifier(
    random_state=42
)

gb_model.fit(X, y)

gb_predictions = gb_model.predict(X)

print(gb_predictions)
```

```python
gb_accuracy = accuracy_score(
    y,
    gb_predictions
)

print("Gradient Boosting Accuracy:")
print(gb_accuracy)
```

## Theory: Gradient Boosting

Gradient Boosting builds models sequentially by minimizing prediction errors.

The algorithm:

1. Creates an initial model.
2. Calculates prediction errors.
3. Builds a new model to correct errors.
4. Repeats the process iteratively.

Popular boosting algorithms include:

- Gradient Boosting
- XGBoost
- LightGBM
- CatBoost

These methods are widely used in predictive analytics and machine learning competitions.

## Theory: Bias-Variance Tradeoff

Machine learning models must balance:

### Bias

Bias refers to errors caused by overly simplistic assumptions.

High bias often leads to underfitting.

### Variance

Variance refers to sensitivity to training data.

High variance often leads to overfitting.

The objective is to find an appropriate balance between bias and variance to maximize generalization performance.

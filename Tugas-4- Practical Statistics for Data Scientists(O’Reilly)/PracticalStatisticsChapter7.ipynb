## Theory: Clustering

Clustering is the process of grouping observations based on similarity.

Observations within the same cluster should be more similar to each other than to observations in other clusters.

Applications include:

- Customer Segmentation
- Market Analysis
- Image Segmentation
- Recommendation Systems

Clustering helps reveal hidden structures within datasets.

```python
import pandas as pd

data = pd.DataFrame({
    'Income': [20,22,25,30,80,85,90,95],
    'Spending': [30,35,28,40,90,88,95,100]
})

data
```

## Theory: K-Means Clustering

K-Means is one of the most popular clustering algorithms.

Procedure:

1. Choose the number of clusters (K).
2. Initialize cluster centroids.
3. Assign observations to the nearest centroid.
4. Update centroids.
5. Repeat until convergence.

The objective is to minimize within-cluster variation.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=2,
    random_state=42,
    n_init=10
)

data['Cluster'] = kmeans.fit_predict(
    data[['Income', 'Spending']]
)

data
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(
    data['Income'],
    data['Spending'],
    c=data['Cluster']
)

plt.title("K-Means Clustering")
plt.xlabel("Income")
plt.ylabel("Spending")

plt.show()
```

## Code Explanation

The K-Means algorithm groups customers into clusters based on income and spending behavior.

Each observation is assigned to the nearest cluster centroid.

The resulting clusters help identify different customer segments and support business decision-making.

## Discussion

K-Means clustering is a simple yet powerful unsupervised learning technique.

By grouping similar observations together, organizations can better understand customer behavior, market trends, and hidden data structures.

Clustering serves as an important exploratory tool in data science.

## Theory: Hierarchical Clustering

Hierarchical Clustering builds a hierarchy of clusters.

Unlike K-Means, it does not require specifying the number of clusters in advance.

Two approaches exist:

### Agglomerative Clustering

Starts with individual observations and gradually merges clusters.

### Divisive Clustering

Starts with one cluster and recursively splits observations.

Hierarchical clustering is useful for understanding nested relationships among observations.

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(
    n_clusters=2
)

data['HC_Cluster'] = hc.fit_predict(
    data[['Income', 'Spending']]
)

print(data)
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    data['Income'],
    data['Spending'],
    c=data['HC_Cluster']
)

plt.title("Hierarchical Clustering")
plt.xlabel("Income")
plt.ylabel("Spending")

plt.show()
```

## Theory: Dimensionality Reduction

Many datasets contain a large number of variables.

Dimensionality reduction techniques simplify datasets while preserving important information.

Benefits include:

- Faster computation
- Reduced storage requirements
- Improved visualization
- Reduced noise
- Better model interpretability

One of the most widely used dimensionality reduction methods is Principal Component Analysis (PCA).

## Theory: Principal Component Analysis (PCA)

Principal Component Analysis transforms original variables into a smaller set of new variables called principal components.

Characteristics:

- Components are uncorrelated.
- Components capture maximum variance.
- Components reduce dataset complexity.

PCA is frequently used in machine learning and exploratory data analysis.

```python
from sklearn.decomposition import PCA

X = data[['Income', 'Spending']]

pca = PCA(
    n_components=2
)

principal_components = pca.fit_transform(X)

print(principal_components)
```

```python
print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)
```

## Code Explanation

The PCA algorithm transforms the original variables into principal components.

The explained variance ratio indicates how much information is retained by each component.

A higher explained variance ratio means that the component captures a larger portion of the dataset's variability.

```python
plt.figure(figsize=(8,5))

plt.scatter(
    principal_components[:,0],
    principal_components[:,1]
)

plt.title("Principal Component Analysis")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.show()
```

## Theory: Explained Variance

Explained variance measures how much information is preserved by each principal component.

For example:

- Component 1 = 85%
- Component 2 = 10%

Total explained variance = 95%

This indicates that most of the dataset information can be represented using only two components.

Explained variance helps determine how many components should be retained.

## Theory: Model-Based Clustering

Model-Based Clustering assumes that data are generated from a mixture of probability distributions.

Unlike K-Means, which relies on distance measures, model-based approaches use probabilistic models.

Advantages:

- Flexible cluster shapes
- Statistical interpretation
- Probability-based assignments

These methods are widely used in advanced clustering applications.

# Chapter Summary

Clustering is an unsupervised machine learning technique used to group similar observations into clusters. Unlike supervised learning, clustering does not require labeled data.

Clustering algorithms help discover hidden patterns, segment customers, detect anomalies, and organize large datasets.

This chapter introduces several clustering techniques including K-Means, Hierarchical Clustering, and DBSCAN, along with methods for evaluating clustering performance.

## Introduction to Clustering

Clustering aims to divide observations into groups such that observations within the same cluster are more similar to each other than observations in different clusters.

Applications include:

- Customer segmentation
- Market analysis
- Image segmentation
- Document organization
- Anomaly detection

Clustering is considered an unsupervised learning technique because labels are not required.

```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data

print(X.shape)
```

## K-Means Clustering

K-Means is one of the most widely used clustering algorithms.

The algorithm works by:

1. Selecting K cluster centers.
2. Assigning observations to the nearest center.
3. Updating cluster centers.
4. Repeating until convergence.

The value of K must be specified before training.

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

clusters = kmeans.fit_predict(X)

print(clusters[:20])
```

```python
print(kmeans.cluster_centers_)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=clusters
)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means Clustering")

plt.show()
```

K-Means groups observations based on distance to cluster centroids. Each observation is assigned to the nearest cluster center.

## Hierarchical Clustering

Hierarchical Clustering creates a hierarchy of clusters.

Two approaches:

- Agglomerative (bottom-up)
- Divisive (top-down)

Agglomerative clustering starts with individual observations and progressively merges clusters.

```python
from sklearn.cluster import AgglomerativeClustering

hierarchical = AgglomerativeClustering(
    n_clusters=3
)

hier_clusters = hierarchical.fit_predict(X)

print(hier_clusters[:20])
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=hier_clusters
)

plt.title("Hierarchical Clustering")

plt.show()
```

Hierarchical clustering builds nested cluster structures and does not require centroid calculations like K-Means.

## Density-Based Clustering with DBSCAN

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) groups observations according to density.

Advantages:

- Identifies arbitrary cluster shapes.
- Detects noise and outliers.
- Does not require specifying the number of clusters beforehand.

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps=0.5,
    min_samples=5
)

dbscan_clusters = dbscan.fit_predict(X)

print(dbscan_clusters[:20])
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=dbscan_clusters
)

plt.title("DBSCAN Clustering")

plt.show()
```

DBSCAN groups observations based on local density and can identify outliers that do not belong to any cluster.

## Cluster Evaluation Metrics

Evaluating clustering performance is more challenging than supervised learning because true labels are often unavailable.

Common evaluation metrics include:

- Silhouette Score
- Davies-Bouldin Index
- Calinski-Harabasz Score

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(
    X,
    clusters
)

print(score)
```

The Silhouette Score measures how well observations fit within their assigned clusters. Higher values generally indicate better clustering quality.

## Choosing the Right Clustering Algorithm

Selection depends on dataset characteristics.

Use K-Means when:

- Clusters are spherical.
- The number of clusters is known.

Use Hierarchical Clustering when:

- Cluster relationships are important.

Use DBSCAN when:

- Outliers are present.
- Cluster shapes are irregular.

## Advanced Clustering Techniques

More advanced clustering methods include:

- Gaussian Mixture Models (GMM)
- Spectral Clustering
- Mean Shift
- Affinity Propagation

These methods can capture more complex structures than traditional clustering algorithms.

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(
    n_components=3,
    random_state=42
)

gmm_clusters = gmm.fit_predict(X)

print(gmm_clusters[:20])
```

```python
exercise_kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

exercise_clusters = exercise_kmeans.fit_predict(X)

print(exercise_clusters[:10])
```

```python
print(
    "K-Means Clusters:",
    len(set(clusters))
)

print(
    "DBSCAN Clusters:",
    len(set(dbscan_clusters))
)
```

```python
print(
    "GMM Clusters:",
    len(set(gmm_clusters))
)
```

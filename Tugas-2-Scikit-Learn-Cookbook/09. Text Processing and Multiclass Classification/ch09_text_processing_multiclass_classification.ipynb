## Introduction to Text Processing

Text processing involves converting raw textual information into structured numerical representations that machine learning algorithms can understand.

Common preprocessing tasks include:

- Lowercasing text
- Removing punctuation
- Removing stop words
- Tokenization
- Stemming
- Lemmatization

Proper preprocessing improves feature quality and model performance.

```python
import pandas as pd

texts = [
    "Machine learning is powerful",
    "Scikit-learn makes machine learning easier",
    "Python is widely used for data science"
]

df = pd.DataFrame({
    "Text": texts
})

df
```

## Text Vectorization Techniques

Machine learning algorithms require numerical inputs.

Text vectorization converts text into numerical feature vectors.

Popular techniques include:

- Count Vectorization
- TF-IDF Vectorization
- Word Embeddings

This chapter focuses on CountVectorizer and TF-IDF.

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X_count = vectorizer.fit_transform(texts)

print(vectorizer.get_feature_names_out())
print(X_count.toarray())
```

CountVectorizer converts text into a matrix of word frequencies. Each column represents a unique word, while each row represents a document.

## TF-IDF Vectorization

TF-IDF (Term Frequency-Inverse Document Frequency) measures the importance of words within a collection of documents.

Words appearing frequently within a document but rarely across all documents receive higher scores.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

X_tfidf = tfidf.fit_transform(texts)

print(tfidf.get_feature_names_out())
print(X_tfidf.toarray())
```

TF-IDF reduces the influence of common words and emphasizes terms that carry more information.

## Feature Extraction from Text

Feature extraction transforms text into meaningful numerical representations.

Common approaches include:

- Bag of Words
- TF-IDF
- N-grams
- Embeddings

N-grams capture relationships between neighboring words.

```python
bigram_vectorizer = CountVectorizer(
    ngram_range=(1, 2)
)

X_bigram = bigram_vectorizer.fit_transform(texts)

print(
    bigram_vectorizer.get_feature_names_out()
)
```

Bigrams consider pairs of consecutive words, allowing models to capture additional contextual information.

## Implementing Text Classification Models

After vectorization, machine learning models can be trained on text features.

Common algorithms include:

- Naive Bayes
- Logistic Regression
- Support Vector Machines
- Random Forests

```python
documents = [
    "I love machine learning",
    "Python is amazing",
    "Artificial intelligence is powerful",
    "Football is exciting",
    "Basketball is popular",
    "Sports improve fitness"
]

labels = [
    "Technology",
    "Technology",
    "Technology",
    "Sports",
    "Sports",
    "Sports"
]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

text_model = Pipeline([
    (
        "vectorizer",
        TfidfVectorizer()
    ),
    (
        "classifier",
        LogisticRegression()
    )
])

text_model.fit(
    documents,
    labels
)

prediction = text_model.predict([
    "Machine learning with Python"
])

print(prediction)
```

The pipeline automatically transforms text into TF-IDF features before applying Logistic Regression for classification.

## Multiclass Classification Strategies

Many text classification tasks involve more than two classes.

Scikit-learn supports multiclass classification using:

- One-vs-Rest (OvR)
- One-vs-One (OvO)
- Multinomial Classification

These approaches extend binary classifiers to multiclass problems.

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

iris = load_iris()

X = iris.data
y = iris.target

multi_model = LogisticRegression(
    max_iter=1000
)

multi_model.fit(X, y)

print(
    multi_model.predict(
        X[:5]
    )
)
```

## Evaluating Text Models

Text classification models should be evaluated using multiple performance metrics.

Common metrics include:

- Accuracy
- Precision
- Recall
- F1 Score
- Confusion Matrix

```python
from sklearn.metrics import classification_report

predictions = text_model.predict(
    documents
)

print(
    classification_report(
        labels,
        predictions
    )
)
```

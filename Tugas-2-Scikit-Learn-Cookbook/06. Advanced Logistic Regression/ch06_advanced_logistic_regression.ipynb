## Overview of Logistic Regression

Logistic Regression is a supervised machine learning algorithm used for classification.

Unlike Linear Regression, Logistic Regression predicts probabilities using the sigmoid function.

The sigmoid function maps values into the range:

0 to 1

This allows predictions to be interpreted as probabilities for classification tasks.

```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target

print(X.shape)
```

## Binary Logistic Regression

Binary classification involves predicting one of two possible classes.

Examples include:

- Spam vs Non-Spam
- Fraud vs Non-Fraud
- Positive vs Negative

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_binary = X[y != 2]
y_binary = y[y != 2]

X_train, X_test, y_train, y_test = train_test_split(
    X_binary,
    y_binary,
    test_size=0.2,
    random_state=42
)

model = LogisticRegression()

model.fit(X_train, y_train)

predictions = model.predict(X_test)

print(predictions[:10])
```

The model learns the relationship between input features and class labels using the logistic function.

Predictions are generated based on estimated class probabilities.

## Multiclass Classification Techniques

Many real-world datasets contain more than two classes.

Scikit-learn supports multiclass classification through:

- One-vs-Rest (OvR)
- Multinomial Logistic Regression

These approaches extend Logistic Regression beyond binary problems.

```python
multiclass_model = LogisticRegression(
    max_iter=1000
)

multiclass_model.fit(X_train, y_train)

print(
    multiclass_model.predict(
        X_test[:5]
    )
)
```

```python
probabilities = multiclass_model.predict_proba(
    X_test[:5]
)

print(probabilities)
```

The predict_proba() method returns class probabilities for each observation.

The class with the highest probability becomes the final prediction.

## Regularization in Logistic Regression

Regularization helps prevent overfitting by controlling model complexity.

Common regularization methods:

- L1 Regularization (Lasso)
- L2 Regularization (Ridge)

Regularization strength is controlled by the C parameter.

```python
regularized_model = LogisticRegression(
    penalty="l2",
    C=1.0,
    max_iter=1000
)

regularized_model.fit(
    X_train,
    y_train
)

print(
    regularized_model.score(
        X_test,
        y_test
    )
)
```

## Multilabel Classification Concepts

In multilabel classification, a single observation may belong to multiple classes simultaneously.

Examples:

- Movie genres
- Document tags
- Product categories

Multilabel classification differs from multiclass classification because multiple labels can be assigned to a single sample.

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.datasets import make_multilabel_classification

X_multi, y_multi = make_multilabel_classification(
    n_samples=200,
    n_features=10,
    n_classes=3,
    random_state=42
)

multi_model = OneVsRestClassifier(
    LogisticRegression(max_iter=1000)
)

multi_model.fit(
    X_multi,
    y_multi
)

predictions = multi_model.predict(
    X_multi[:5]
)

print(predictions)
```

OneVsRestClassifier trains a separate classifier for each label and combines their predictions.

This approach allows Logistic Regression to handle multilabel datasets.

## Model Evaluation Metrics

Classification models should be evaluated using appropriate metrics.

Common metrics include:

- Accuracy
- Precision
- Recall
- F1 Score
- ROC-AUC

Each metric provides different insights into model performance.

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(
    y_test,
    multiclass_model.predict(X_test)
)

print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report

print(
    classification_report(
        y_test,
        multiclass_model.predict(X_test)
    )
)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(
    y_test,
    multiclass_model.predict(X_test)
)

print(cm)
```

```python
exercise_model = LogisticRegression(
    C=0.5,
    max_iter=1000
)

exercise_model.fit(
    X_train,
    y_train
)

print(
    exercise_model.score(
        X_test,
        y_test
    )
)
```

```python
multiclass_score = multiclass_model.score(
    X_test,
    y_test
)

print(multiclass_score)
```

```python
sample_probabilities = multiclass_model.predict_proba(
    X_test[:3]
)

print(sample_probabilities)
```

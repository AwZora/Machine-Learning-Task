Machine learning models rely heavily on the quality of the input data.

Poor-quality data may cause:

- Reduced accuracy
- Overfitting
- Underfitting
- Biased predictions
- Poor generalization

Therefore, preprocessing is essential before model training.

```python
import pandas as pd
import numpy as np

data = {
    "Age": [25, 30, np.nan, 40, 35],
    "Salary": [5000, 7000, 6000, np.nan, 8000]
}

df = pd.DataFrame(data)

df
```

## Cleaning and Preparing Data

Data cleaning aims to improve data quality before training.

Common tasks include:

- Removing duplicates
- Fixing incorrect values
- Standardizing formats
- Handling missing values

```python
duplicate_df = pd.DataFrame({
    "A": [1, 1, 2, 2, 3],
    "B": [5, 5, 6, 6, 7]
})

print("Before:")
print(duplicate_df)

duplicate_df = duplicate_df.drop_duplicates()

print("\nAfter:")
print(duplicate_df)
```

## Handling Missing Data

Missing values are among the most common problems in machine learning datasets.

Several strategies can be used:

- Remove rows
- Remove columns
- Mean imputation
- Median imputation
- Most frequent value imputation

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="mean")

imputed_data = imputer.fit_transform(df)

print(imputed_data)
```

The SimpleImputer replaces missing numerical values with the mean value of the corresponding feature.

## Scaling Techniques

Machine learning algorithms often perform better when features are on similar scales.

Common scaling methods:

- StandardScaler
- MinMaxScaler
- RobustScaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(imputed_data)

print(scaled_data[:5])
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()

normalized_data = minmax.fit_transform(imputed_data)

print(normalized_data[:5])
```

## Encoding Categorical Variables

Machine learning algorithms generally require numerical input. Therefore, categorical variables must be converted into numerical representations before model training.

Common encoding methods include:

- Label Encoding
- One-Hot Encoding
- Ordinal Encoding

The choice of encoding method depends on the nature of the categorical variable and the machine learning algorithm being used.

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

colors = pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Red", "Blue"]
})

encoder = LabelEncoder()

colors["Encoded"] = encoder.fit_transform(colors["Color"])

print(colors)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(colors[["Color"]])

print(encoded)
```

Label Encoding assigns a unique integer to each category. One-Hot Encoding creates separate binary columns for each category and is generally preferred when categories do not have an inherent order.

## Introduction to Pipelines in scikit-learn

As machine learning projects grow larger, preprocessing and modeling steps can become difficult to manage.

Pipelines provide a structured approach to combine multiple operations into a single workflow.

Benefits of pipelines:

- Cleaner code
- Improved reproducibility
- Prevention of data leakage
- Easier hyperparameter tuning

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target

pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="mean")),
    ("scaler", StandardScaler()),
    ("classifier", LogisticRegression(max_iter=1000))
])

pipeline.fit(X, y)

print("Pipeline Accuracy:", pipeline.score(X, y))
```

## Visualizing Pipelines

Understanding pipeline structure is important for debugging and documentation purposes.

A pipeline can be inspected by printing its structure.

```python
print(pipeline)
```

The printed output shows the sequence of operations executed within the pipeline. Each preprocessing step is applied before the final machine learning model.

## Feature Engineering

Feature engineering is the process of creating new features or modifying existing features to improve model performance.

Good features often contribute more to model performance than complex algorithms.

Common feature engineering techniques include:

- Polynomial features
- Feature interactions
- Date feature extraction
- Aggregation features
- Domain-specific transformations

```python
from sklearn.preprocessing import PolynomialFeatures

X_small = [[2, 3],
           [4, 5],
           [6, 7]]

poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(X_small)

print(X_poly)
```

PolynomialFeatures generates additional features based on mathematical combinations of existing features. These new features can help linear models capture non-linear relationships.

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

exercise_data = pd.DataFrame({
    "Age": [20, 25, np.nan, 30, 35],
    "Income": [3000, np.nan, 4500, 5000, 5500]
})

imputer = SimpleImputer(strategy="mean")

result = imputer.fit_transform(exercise_data)

print(result)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_result = scaler.fit_transform(result)

print(scaled_result)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    iris.data,
    iris.target,
    test_size=0.2,
    random_state=42
)

pipeline.fit(X_train, y_train)

accuracy = pipeline.score(X_test, y_test)

print("Test Accuracy:", accuracy)
```
